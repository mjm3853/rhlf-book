```python
import pandas as pd

# Create a DataFrame from the mock data
df = pd.DataFrame({
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05'],
    'Sales': [100, 120, 110, 130, 140],
    'Profit': [10, 12, 11, 13, 14]
})

# Print the DataFrame to verify it's correct
print(df)

# Now, let's visualize the mock data with seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))

sns.lineplot(data=df, x='Date', y='Sales')
sns.lineplot(data=df, x='Date', y='Profit')

plt.title('Mock Data Visualization')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()

# Print a message to confirm the visualization was successful
print("Mock data successfully visualized with seaborn.")
```
